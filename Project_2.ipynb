## Fluctuations Of Dollar to Euro Exchange Rates (2001-2023)

```python
import pandas as pd
```

```python
exchange_rates = pd.read_csv("euro-daily-hist_1999_2022.csv")
exchange_rates.shape
```

```python
exchange_rates.head()
```

```python
exchange_rates.info()
```

```python
exchange_rates.nunique()
```

```python
exchange_rates.rename(columns = {'[US dollar ]' : 'US_dollar','Period\\Unit:' : 'Time'},inplace=True)
```

```python
exchange_rates.head()
```

```python
exchange_rates["Time"].info()
```

```python
exchange_rates["Time"] = pd.to_datetime(exchange_rates["Time"])
```

```python
exchange_rates.sort_values('Time', inplace = True)
```

```python
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time','US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts() 
```

```python
euro_to_dollar=euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

```python
import matplotlib.pyplot as plt
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

```python
values = pd.DataFrame()
values["daily_values"] = pd.Series(range(1,20,2))
values
```

```python
values["rolling_mean_2"] = values["daily_values"].rolling(2).mean()
values
```

```python
values["rolling_mean_3"] = values["daily_values"].rolling(3).mean()
values["rolling_mean_5"] = values["daily_values"].rolling(5).mean()
values
```

```python
plt.figure(figsize=(16,13))
plt.subplot(3,2,1)
plt.plot(euro_to_dollar["Time"],euro_to_dollar["US_dollar"])
plt.title('Original Values',weight='bold')
plt.ylim(0.8,1.7)
for i, rolling_mean in zip([2,3,4,5,6,],[7,30,50,100,365]):
     plt.subplot(3,2,i)
     plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
     plt.title('Rolling Window:' + str(rolling_mean),weight='bold')
     plt.ylim(0.8,1.7)
plt.tight_layout() # Auto_adjust the padding between the subplots
plt.show()
```

- As the time period grows the pattern becomes more identifiable

```python
euro_to_dollar["rolling_mean"]=euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

```python
financial_crisis = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2006) & (euro_to_dollar['Time'].dt.year <= 2009)]
financial_crisis_7_8 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2007) & (euro_to_dollar['Time'].dt.year <= 2008)]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

# Adding the plot
fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(financial_crisis['Time'], financial_crisis['rolling_mean'], linewidth=2, color='#FF5733')  # Change line color
ax.plot(financial_crisis_7_8['Time'], financial_crisis_7_8['rolling_mean'], linewidth=2, color='#0066cc')  # Change line color

# Customize x-axis tick labels
ax.set_xticklabels([])
x = 0.02
for year in ["2006", "2007", "2008", "2009", "2010"]:
    ax.text(x, -0.08, year, alpha=0.5, fontsize=11, transform=plt.gca().transAxes)
    x += 0.22888

# Customize y-axis tick labels
ax.set_yticklabels([])
y = 0.07
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(-0.04, y, rate, alpha=0.5, fontsize=11, transform=plt.gca().transAxes)
    y += 0.2333

# Adding a title and a subtitle
ax.text(-0.05, 1.2, "Euro_usd rate reached 1.59 during 2007_2008 financial crisis", weight='bold',
        transform=plt.gca().transAxes)
ax.text(-0.05, 1.1, "Euro_usd exchange rate between 2006 and 2010", size=12, transform=plt.gca().transAxes)

# Adding a signature
ax.axvspan(xmin=pd.to_datetime("2008-4-1"), xmax=pd.to_datetime("2008-9-1"), ymin=0.09, alpha=0.3, color='grey')

plt.show()
```

`This visual representation illustrates the trajectory of the Euro to US dollar exchange rate spanning the years 2006 to 2010, with a specific focus on the tumultuous financial crisis period in 2007-2008. Commencing in 2006, the Euro exhibits a strengthening trend against the dollar, denoted by the transition from an orange line to a blue one as the exchange rate ascends. The pinnacle is marked at 1.59 during the crisis, discernible by the shaded region, symbolizing the zenith of the economic downturn. Following this peak, there is a discernible and abrupt descent, capturing the climax of the crisis's impact. In the aftermath of the crisis, the exchange rate demonstrates stabilization and initiates a recovery phase, evident in the line reverting to an orange hue. This shift suggests a period of economic recuperation or stabilization extending towards the conclusion of 2010.`

```python
corona_crisis_20=euro_to_dollar.loc[(euro_to_dollar['Time']>="2020-01-01")&(euro_to_dollar['Time']<="2020-12-31")]
corona_crisis=euro_to_dollar.loc[(euro_to_dollar['Time']>="2016-01-01")&(euro_to_dollar['Time']<="2019-12-31")]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

# Adding the plot
fig, ax = plt.subplots(figsize=(9, 3))
ax.plot(corona_crisis['Time'], corona_crisis['rolling_mean'], linewidth=2, color='#FF5733')  # Change line color
ax.plot(corona_crisis_20['Time'], corona_crisis_20['rolling_mean'], linewidth=3, color='#4CBB17')  # Change line color

# Customize x-axis tick labels
ax.set_xticklabels([])
x = 0.02
for year in ["2016", "2017", "2018", "2019", "2020"]:
    ax.text(x, -0.08, year, alpha=0.5, fontsize=11, transform=plt.gca().transAxes)
    x += 0.22888

# Customize y-axis tick labels
ax.set_yticklabels([])
y = 0.02
for rate in ['1.05', '1.10', '1.15', '1.20']:
    ax.text(-0.04, y, rate, alpha=0.5, fontsize=11, transform=plt.gca().transAxes)
    y += 0.248

# Adding a title and a subtitle
ax.text(-0.05, 1.2, "Euro_usd rate reached 1.22 during 2020's COVID-19 crisis", weight='bold',
        transform=plt.gca().transAxes)
ax.text(-0.05, 1.1, "Euro_usd exchange rate between 2016 and 2020", size=12, transform=plt.gca().transAxes)

plt.show()
```

`This visual representation depicts the dynamics of the Euro-to-US dollar exchange rate from 2016 to 2020. The graph unveils a general descending pattern in the Euro's value relative to the dollar, persisting until the conclusion of 2019. However, a distinctive upswing is evident in 2020, marked by a notable peak at 1.22 amid the COVID-19 crisis, denoted by the transition in line color from a rosy hue to a cerulean shade. This apex likely mirrors the market's response to the economic repercussions of the pandemic, where diverse factors such as economic strategies, investor sentiment, and global events can collectively influence currency valuations`

```python
bush_obama_trump = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year <= 2021)]
bush = bush_obama_trump.copy()[bush_obama_trump['Time'].dt.year < 2009]
obama = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >=2009) & (bush_obama_trump['Time'].dt.year <= 2017)]
trump = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year <= 2021)]
```

```python
# Adding the FiveThirtyEight Style
style.use('fivethirtyeight')

# Adding the subplots
plt.figure(figsize=(14, 8))
# Pattern 1
ax1 = plt.subplot(3, 3, 1)  # row, col, index
ax2 = plt.subplot(3, 3, 2)
ax3 = plt.subplot(3, 3, 3)
# Pattern 2
ax4 = plt.subplot(3, 1, 2)  # row, col, index
axes = [ax1, ax2, ax3, ax4]

# Change to all the subplots
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6$'], alpha=0.4)

# Ax1: Bush
ax1.plot(bush['Time'], bush['rolling_mean'], color='#FF1493')  # Deep Pink
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '', '2007', '', '2009'], alpha=0.3, size=12)
ax1.text(0.099, 2.45, 'BUSH', fontsize=18, weight='bold', color='#FF1493', transform=plt.gca().transAxes)
ax1.text(0.08, 2.34, '(2001-2009)', weight='bold', alpha=0.3, transform=plt.gca().transAxes)

# Ax2: Obama
ax2.plot(obama['Time'], obama['rolling_mean'], color='#00FF7F')  # Spring Green
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '', '2015', '', '2017'], alpha=0.3)
ax2.text(0.45, 2.45, 'OBAMA', fontsize=18, weight='bold', color='#00FF7F', transform=plt.gca().transAxes)
ax2.text(0.44, 2.34, '(2009-2017)', weight='bold', alpha=0.3, transform=plt.gca().transAxes)

# Ax3: Trump
ax3.plot(trump['Time'], trump['rolling_mean'], color='#0000FF')  # Blue
ax3.set_xticklabels(['', '2017', '', '2018', '', '2019', '', '2020', '', '2021'], alpha=0.3)
ax3.text(0.818, 2.45, 'TRUMP', fontsize=18, weight='bold', color='#0000FF', transform=plt.gca().transAxes)
ax3.text(0.808, 2.34, '(2017-2021)', weight='bold', alpha=0.3, transform=plt.gca().transAxes)

# Ax4: Bush-Obama_Trump
ax4.plot(bush['Time'], bush['rolling_mean'], color='#FF1493')
ax4.plot(obama['Time'], obama['rolling_mean'], color='#00FF7F')
ax4.plot(trump['Time'], trump['rolling_mean'], color='#0000FF')
ax4.set_xticks([])

# Adding a title and a subtitle
ax1.text(-0.05, 2.8, 'EURO-USD rate averaged 1.22 under the last three US presidents', fontsize=20,
         weight='bold', transform=plt.gca().transAxes)
ax1.text(-0.05, 2.55, '''EURO-USD exchange rates under George W. Bush (2001-2009), Barack Obama (2009
''', fontsize=14, transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()
```

`George W. Bush (2001-2009) :`
During the leadership of President George W. Bush, the EURO-USD exchange rate witnessed notable fluctuations. Commencing at approximately parity (1.0) in 2001, the rate exhibited a consistent ascent in the subsequent years, culminating at just above 1.6 around 2008. This era marked a phase of the Euro gaining strength against the dollar, signifying an increase in the cost of euros in terms of dollars. The conclusion of Bush's tenure was characterized by a sharp downturn in the rate, aligning with the global financial crisis of 2007-2008, which profoundly impacted the value of the USD.

`Barack Obama (2009-2017) :`
President Barack Obama assumed office immediately following the financial crisis, with the exchange rate having already descended from its peak. Throughout his presidency, the exchange rate mainly fluctuated within the range of 1.2 to 1.4. Despite occasional fluctuations, the overall trajectory suggests a gradual decline, indicating either a strengthening of the dollar against the euro or a weakening of the euro against the dollar during this period. As his term concluded, the exchange rate hovered slightly above 1.0.

`Donald Trump (2017-2021) :`
Under the administration of President Donald Trump, the EURO-USD exchange rate exhibited a more stable trend, characterized by fewer pronounced fluctuations than in the preceding periods. It remained relatively stable, fluctuating roughly between 1.1 and 1.2. This stability indicated reduced volatility and a narrower range of rate fluctuation. Towards the latter years of his term, a subtle upward trend emerged, suggesting a mild weakening of the USD or a strengthening of the euro.
